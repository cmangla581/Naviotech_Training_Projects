Key symptoms inlcude the chest pain (angina), shortness of breath, dizziness and discomfort in arms, back or jaw. Major risks include the high blood pressure, smoking, diabetes and obesity .

Common types of heart disease include : 

1. Coronary Artery Disease 
2. Heart Failure 
3. Arrhythmia 
4. Heart Valve Disease

Now, we will be splitting the  dataset into the features and labels which means the dependent and independent features. 

```python
x = data.drop(['heart disease'], axis = 1) 
y = data.filter(['heart disease'], axis = 1) 
```

```python
x  
# These are all the features on which the predictions will be made 
```

```python
y 
# This is the feature which is basicall the result  
```

```python
# Now comes the part of train, testing and splittting the data 

from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size = 0.20) 
```

```python
x_train # Data on which model is trained 
```

```python
x_test   # Data on which model is tested 
```

```python
y_train  # Data on which model is trained 
```

```python
y_test  # Data on which model is tested 
```

```python
# Now comes the important step of preprocessing the data and scaling the data by applying the Scaling feature to scale thw data 

from sklearn.preprocessing import StandardScaler 

sc = StandardScaler()  

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)  
```

Now, we will import the  svm library for the model training and classification of the outcome of the presence or absence of the disease. 

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression() 
classifier = model.fit(x_train, y_train) 
classifier  
```

```python
# Now, we make the further analysis in the model  by the predictions and probability predictions 

y_pred = classifier.predict(x_test) 
y_pred 

# 2 stands for Heart Disease and 1 stands for No Heart Disease 
```

```python
# Now, we will perform the analysis of the model by the probability  

y_prob = classifier.predict_proba(x_test)[:,1] 
y_prob 
```

Now, we will evaluate the model by evaluating the metrics like acccuracy score, confusion matrix and classification report. 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report  

print("Accuracy Score: ", accuracy_score(y_test, y_pred))  

print("\nClassification Report:\n ", classification_report(y_test, y_pred)) 

print("\nConfusion Matrix:\n",confusion_matrix(y_test,y_pred)) 
```

```python
import pandas as pd 

example = pd.DataFrame({
    'age': [60,45],
    'sex ': [1,0],
    'chest pain type': [1,2],
    'resting blood pressure': [180,120],
    'serum cholestoral': [300, 120],
    'fasting blood sugar': [1,0],
    'resting electrocardiographic results': [2,0],
    'max heart rate': [180,110],
    'exercise induced angina': [1,0],
    'oldpeak': [2.5, 1.0],
    'ST segment': [2,3],
    'major vessels': [3,0],
    'thal': [7,3]
}) 

example
```
