# Filling the missing values 
Now, we will be filling values and here e will be using the fillna value and not dropna, as because the dropna value removes the whole row with missing vallues and hence there is a loss of large dataset. 

```python
# Handle the missing values 

# Filling the categorical data with mode 

cat_cols = data.select_dtypes(include = "object").columns 
for col in cat_cols: 
    data[col].fillna(data[col].mode()[0], inplace = True) 

num_cols = data.select_dtypes(include = np.number).columns 
for col in num_cols: 
    data[col].fillna(data[col].median(), inplace = True) 
```

```python
data  
# Here, the data has been implanted and the missing values have been filled 
```

Here, Loan_ID and the Gender are un necessary columns and hence, they have to be removed for the data preprocessing. 

```python
data2 =  data.drop(["Loan_ID", "Gender"], axis = 1)  
data2  

# Noow, all the pre processing has to be ddone from the data 
```

```python
print(data2["Married"].unique()) 
print(data2["Dependents"].unique()) 
print(data2["Education"].unique()) 
print(data2["Self_Employed"].unique()) 
print(data2["Credit_History"].unique()) 
print(data2["Property_Area"].unique()) 
print(data2["Loan_Status"].unique()) 
```

Now, we will convert the categorical data into the numerical data for the further preprocessing. 

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder() 

for col in data2.select_dtypes(include = "object"):  
    data2[col] = le.fit_transform(data2[col]) 
```

Now, comes dividing into  the train and test splits

```python
# First, we will be dividing into the features and labels 

x = data2.drop(["Loan_Status"], axis = 1) 
x  

# this is the data, on which the predictions will be made 
```

```python
y = data2.filter(["Loan_Status"], axis = 1) 
y 
```

Now, we will be dividing the whole dataset into the training and the test features. 

```python
from sklearn.model_selection import train_test_split  
x_train,x_test, y_train, y_test = train_test_split(x,y, random_state = 42, test_size = 0.2) 
```

```python
x_train  
# Data on which the model is trained 
```

```python
x_test 
# Here, this is the data on which the model is tested 
```

```python
y_train 
# Data on whcih the model is trained 
```

```python
y_test 
# Data on which the model is tested 
```

```python
# Now, we will perfrom the standard scaling of the hole dataset 
from sklearn.preprocessing import StandardScaler 

sc = StandardScaler() 

x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test) 
```

Now, comes the main part of implementing the Ml model of KNeighborsClassifier to make the further predictions and for training the model. 

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 7)  

classifier = model.fit(x_train, y_train) 
classifier 
```

```python
# Now, we will be making the predcitoons, to carry out the further pre processing.  

y_pred = classifier.predict(x_test)
y_pred 
```

```python
# Now, we will making the probabilites prediction to carry out the further preprocessing as: 
y_prob = classifier.predict_proba(x_test)[:,1]
y_prob  

# Here, the probabilities have been calculated 
```

Here, as we can see that the predicctions have been made and hence, now we will be calculating the metrics to carry out the further preprocessing. 

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 

print("Acuracy Score:\n",accuracy_score(y_test, y_pred)) 
print("Classification Report:\n", classification_report(y_test, y_pred)) 
print("Confusion Matrix:\n",confusion_matrix(y_test, y_pred))  
```

Now, comes the part of the plotting of the histograms and the box plots  to make the project visually more understandable.  

```python
def plot_histogram(data2, column_name): 
    plt.figure(figsize = (5,3)) 
    sns.histplot(data2[column_name], kde = True) 
    plt.title(f"Distribution of {column_name}") 

    # Calculate the mean and the mmedian values of the columns 
    col_mean = data2[column_name].mean()
    col_median = data2[column_name].median() 

    # Adding of the vartial lines for the mean and median 
    plt.axvline(col_mean, color = "red", linestyle = "--", label = "Mean") 
    plt.axvline(col_median, color = "green", linestyle = "-", label = "Median")

    plt.legend()
    plt.show() 
```
